```python
import subprocess
def get_response(input_text: str, agent_path:str ):
    # 啟動 exe
    proc = subprocess.Popen(
        [agent_path],
        stdin=subprocess.PIPE,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    # 將 input_text 寫入 stdin，並讀取 stdout/stderr
    stdout, stderr = proc.communicate(input_text)

    return stdout, stderr

def get_move(response_text):
    return response_text.split("\n")[-2]

def get_FEN(response_text):
    return response_text.split("\n")[-1]

def get_board(response_text):
    board = ''
    for s in  response_text.split("\n")[-14:]:
        board += s+"\n"
    return board
```

```python
import random
def gen_board():
    pool = 'EenCPKAr/PaeCErpc/ApanpcPN/pkPpNRRP'.replace('/', '')
    pool = list(pool)
    random.shuffle(pool)
    board = ''
    for s in pool:
        board += s
    board = board[:8] + '/' + board[8:16] + '/' + board[16:24] + '/' + board[24:32]
    # board += ' r'
    rd = random.randint(0, 1)
    color = [ ' r', ' b' ]
    board += color[rd]
    return board
```

```python
# board = "A6p/8/8/8 b"
from IPython.display import display, clear_output
import time
def play(show_content = False):
    board = gen_board()
    Red_Win = False
    Black_Win = False
    player1 = './wakasagi'
    player2 = './agents/MCS_MyAGENT'
    move = 0

    players = {
        'r': player1,
        'b': player2
    }
    first = board[-1]
    current_player = players[first]

    while not ( Red_Win or Black_Win):
        res = (get_response(board, current_player)[0])
        board = get_FEN(res)
        b = board.split(" ")[0]
        Black_Win = not any(c.islower() for c in b)
        Red_Win = not any(c.isupper() for c in b)
        move += 1
        if show_content:
            clear_output(wait=True)

            print(first, 'first')
            print('red:', players['r'])
            print('black:', players['b'])
            print(move, end='\r')
            print(get_board(res))
            # time.sleep(0.5)

        if Black_Win or Red_Win:
            break

        current_player = players[ board[-1] ]

    if Black_Win:
        return players['b']
    else:
        return players['r']
```
