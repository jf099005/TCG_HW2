```python
import subprocess
def get_response(input_text: str, agent_path:str ):
    # 啟動 exe
    proc = subprocess.Popen(
        [agent_path],
        stdin=subprocess.PIPE,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    # 將 input_text 寫入 stdin，並讀取 stdout/stderr
    stdout, stderr = proc.communicate(input_text)

    return stdout, stderr

def get_move(response_text):
    return response_text.split("\n")[-2]

def get_FEN(response_text):
    return response_text.split("\n")[-1]

def get_board(response_text):
    board = ''
    for s in  response_text.split("\n")[-14:]:
        board += s+"\n"
    return board
```

```python
# board = "A6p/8/8/8 b"
from IPython.display import display, clear_output
import time
board = 'EenCPKAr/PaeCErpc/ApanpcPN/pkPpNRRP r'
Red_Win = False
Black_Win = False

move = 0

show_content = False

while not ( Red_Win or Black_Win):
    res = (get_response(board,"./wakasagi")[0])
    board = get_FEN(res)
    b = board.split(" ")[0]
    Black_Win = not any(c.islower() for c in b)
    Red_Win = not any(c.isupper() for c in b)
    move += 1
    clear_output(wait=True)
    print(move, end='\r')
    if show_content:
        print(get_board(res))
        time.sleep(0.5)

    if Black_Win or Red_Win:
        break
print('\n')
print('board:', board)
print("winner:")
if Red_Win:
    print("\t Red")
else:
    print("\t Black")

print('move:', move)
```
